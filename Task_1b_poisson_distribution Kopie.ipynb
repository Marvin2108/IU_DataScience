```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([])
x = np.arange(0,32)

for i in range(0,32):
    y = np.append(y,(poisson.pmf(i,19))*100)
    
    # Bedingung für P(X>=x)< 0.5%
    if 100-np.sum(y) <= 0.5:
        break
        
#Berechnung Median & Varianz
median = poisson.median(19)
varianz = poisson.var(19)
stabw = poisson.std(19)
print("Median:",median, "Varianz:", varianz, "Standardabweichung", stabw)

#Beschriftung Grafik
plt.title("Poisson-Verteilung für Meteoriteneinschlag \n",fontsize=15)
plt.ylabel("Wahrscheinlichkeit in %")
plt.xlabel("Anzahl der Einschläge k")

# Poisson Verteilung visualisieren
plt.stem(x,y, basefmt=" ", label="Poisson-Verteilung µ = 19")
#plt.grid(True)
plt.xticks(np.arange(0,36,2))
plt.yticks(np.arange(0,11,0.5))

# Visualisierung P(X>=x)< 0.5%
plt.axvline(i,color="red",ymax=8, label="P(X>=x)< 0.5%")

#Median und Varianz visualisieren
plt.bar(median,poisson.pmf(19,19)*100,color='orange',width=0.5, label="µ = Median = Varianz = 19")
plt.arrow(x=median,y=2.0,dx=stabw,dy=0,width=0.3, color="green",label="Standardabweichung")
plt.arrow(x=median,y=2,dx=-stabw,dy=0,width=0.3,color="green")

# Legende
plt.legend(loc="upper left",prop={'size': 8})

plt.show()
```
